```python
def condense_csv(filename , id_name ):
    
    with open(filename, encoding='utf-8') as file:
        rows = csv.reader(file)
        l_values = []
        l_name = []
        dic_id = ""
        dic = {}
        for row in rows:
            if dic.setdefault(row[0],0) == 0:
                if l_values != []:
                    print(l_values)
                    dic[dic_id] = l_values 
                    l_values = []
                dic_id = row[0]  
            
            if l_name.count(row[1]) == 0:
                l_name.append(row[1])    

            l_values.append(row[2])    
        
        dic[dic_id]  =  l_values     
        dic = dict(sorted(dic.items()))
        l_name = [id_name] + l_name
        print(l_name, l_values)
        l=[[k, *v] for k,v in dic.items()]
        with open('C:/Users/Tsvetochek/study/condensed.csv', 'w', encoding='utf-8', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(l_name)           
            writer.writerows(l)

    return writer



condense_csv('C:/Users/Tsvetochek/study/test.txt' , 'ID' )
```
